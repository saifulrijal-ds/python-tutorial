```
# import the required Python package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
# please upload file gapminder-FiveYearData.csv
# read the CSV file to the DataFrame format, and assign it to a variable df
df = pd.read_csv("/content/gapminder-FiveYearData.csv")
```

```
# let's display some rows of df DataFrame

display(df.head())
```

```
# create a new dataframe that is the result of a previous dataframe pivoting.
# column "country" as new index, column "year" as new column, and column
# "lifeExp" populating new frame's value

new_df = df.pivot(index="country", columns="year", values="lifeExp")
new_df
```

```
plt.figure(figsize=(15, 40))

sns.heatmap(new_df, annot=True, fmt=".2f").get_figure().savefig("heatmap.png")
```
